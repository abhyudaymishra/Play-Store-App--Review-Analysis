Finally, we can examine the pricing of apps in the Google Play Store. The majority of apps in the store are free, with a small proportion of paid apps. Paid apps tend to have more features and functionalities than free apps, which may explain why users are willing to pay for them. 

In conclusion, Exploratory Data Analysis (EDA) on the Google Play Store dataset provides valuable insights into the trends and patterns in the Android app market. ## **DATA CLEANING:-**  
Data cleaning is the process of identifying and correcting (or removing) inaccuracies, inconsistencies, and missing values in a dataset. The process can be broken down into several steps:

1. Identify the problem: This step involves reviewing the data and identifying any inaccuracies, inconsistencies, or missing values that need to be corrected.
2. Data cleaning: This step involves making the necessary corrections to the data. This can include removing outliers, filling in missing values, or standardizing data.
3. Data validation: This step involves checking the cleaned data to ensure that it is accurate and complete.
4. Data transformation: This step involves converting the data into a format that is more useful for analysis. This may include converting data into a specific format, creating new variables, or aggregating data.
5. Data loading: This step involves loading the cleaned and transformed data into a database or other storage system for further analysis.

## 1. What is the average rating for apps in the Play Store?

```python
# to calculate the average rating we will use mean method of pandas library 
average_rating = play_store['Rating'].mean()

print("The average rating for apps in the Play Store is", average_rating)
```

```python
#plotting the Bar chart to visulize the average ratings
# convert the average rating to a DataFrame
average_rating_df = pd.DataFrame({'Metric': ['Average Rating'], 'Value': [average_rating]})

# create the bar plot
average_rating_df.plot(x='Metric', y='Value', kind='bar', xlabel='Metric', ylabel='Value', title='Average rating for apps in Play Store', ylim=(1, 5),figsize=(8,8),color='green')
```

## 2. What is the division between free and paid apps? 

```python
# Finding  the number of free apps
free_apps = play_store[play_store['Type'] == 'Free'].count()['Type']

# Calculating the number of paid apps
paid_apps = play_store[play_store['Type'] == 'Paid'].count()['Type']

# Calculating the total number of apps
total_apps = free_apps + paid_apps

#Calculating the proportion of free apps
proportion_free_apps = free_apps / total_apps

# Calculating the proportion of paid apps
proportion_paid_apps = paid_apps / total_apps

# Printing the results
print("Number of free apps:", free_apps)
print("Number of paid apps:", paid_apps)
print("Proportion of free apps:", proportion_free_apps)
print("Proportion of paid apps:", proportion_paid_apps)
```

## 3. Are paid apps rated higher than free apps?

```python
# first we will separate the ratings of paid and free apps
paid_apps = play_store[play_store['Type'] == 'Paid']
free_apps = play_store[play_store['Type'] == 'Free']

#then we will calculate the average rating for each group
paid_apps_average_rating = paid_apps['Rating'].mean()
free_apps_average_rating = free_apps['Rating'].mean()

# now creating the dataframe
compare_ratings = pd.DataFrame({'App Type': ['Paid', 'Free'],
                                'Average Rating': [paid_apps_average_rating, free_apps_average_rating]})

# setting the index
compare_ratings = compare_ratings.set_index('App Type')

# showing the dataframe
print(compare_ratings)
```

## 8. What are the common features of high-rated apps and how do they differ from low-rated apps?
 One approach to answer this question is to compare the app features, such as genre, price, size, and number of installs, between high-rated apps and low-rated apps. Here's a syntax to compare the distribution of features between apps with a rating of 4 or above and apps with a rating below 2:

```python
low_rated = play_store[play_store['Rating'] <2]
low_rated_genres = low_rated['Genres'].value_counts()
last_20_genres=low_rated_genres.head(20)

high_rated = play_store[play_store['Rating'] >= 4]
high_rated_genres = high_rated['Genres'].value_counts()
top_20_genres=high_rated_genres.head(20)

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(30,15))
last_20_genres.plot(kind='bar', ax=axs[0], title='Low Rated Apps by Genre', color="red")
axs[0].set_xlabel('Genres')
axs[0].set_ylabel('Frequency')
axs[0].set_xticklabels(last_20_genres.index, rotation=90)

top_20_genres.plot(kind='bar', ax=axs[1], title='High Rated Apps by Genre', color="green")
axs[1].set_xlabel('Genres')
axs[1].set_ylabel('Frequency')
axs[1].set_xticklabels(top_20_genres.index, rotation=90)

plt.show()
```

## 13. What is the overall sentiment (positive, negative, or neutral) of the user reviews for Top 10 apps in the dataset?

```python
# First we will group by app name and sentiment and find the mean of sentiment polarity
app_sentiment = user_review.groupby(['App', 'Sentiment'], as_index=False)['Sentiment_Polarity'].mean()

# then we will try to pivot the data to have sentiment as columns and values as mean sentiment polarity
app_sentiment_pivot = app_sentiment.pivot(index='App', columns='Sentiment', values='Sentiment_Polarity')

# limit the data to the top 10 apps
top_apps = user_review['App'].value_counts().index[:10]
app_sentiment_pivot = app_sentiment_pivot.loc[top_apps]

#plotting the Heatmap from seaborn library
plt.figure(figsize=(10,8))
sns.heatmap(app_sentiment_pivot, cmap="Wistia", annot=True)
plt.show()
```

## 14.How does the sentiment polarity (the degree to which a sentiment is positive or negative) vary among the user reviews for top 10 apps?

```python
# Extracting the top 10 apps based on the number of reviews
top_10_apps = user_review['App'].value_counts().head(10)

# Filtering the reviews data for the top 10 apps only
top_10_reviews = user_review[user_review['App'].isin(top_10_apps.index.tolist())]
top_10_reviews
```

```python
#to viuslize the sentiment polarity we will try to plot a Violin Chart.
plt.figure(figsize=(14,12))
sns.violinplot(x='App', y='Sentiment_Polarity', data=top_10_reviews)
plt.xticks(rotation=90)
plt.title('Sentiment Polarity Variation Among User Reviews for Top 10 Apps')
plt.show()
```
